### Gold Validation

```python
%%pyspark
from pyspark.sql.functions import col

# Check negative days
df = spark.sql("""
SELECT *
FROM gold.fact_claim
WHERE days_to_submit < 0
""")

if df.count() > 0:
    raise Exception("Validation Failed: Negative days_to_submit found")
else:
    print("Validation Passed: No negative values")
```

```python
%%sql
SELECT *
FROM gold.fact_claim
WHERE claim_amount < 0;
```

```python
df = spark.sql("SELECT COUNT(*) as cnt FROM gold.fact_claim")
count = df.collect()[0]["cnt"]

if count == 0:
    raise Exception("Validation Failed: fact_claims is empty")
else:
    print("Validation Passed: fact_claims has data")
```

```python
df = spark.sql("""
SELECT *
FROM gold.fact_claim
WHERE patient_key IS NULL
   OR provider_key IS NULL
""")

if df.count() > 0:
    raise Exception("Validation Failed: Null foreign keys found")
else:
    print("Validation Passed: No null foreign keys")
```
